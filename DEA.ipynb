````python
import ast

# Step 1: Create a cleaning/normalization function
def normalize_outlet(outlet):
    """Clean and normalize outlet names"""
    if not outlet or not isinstance(outlet, str):
        return None

    # Remove leading/trailing whitespace
    outlet = outlet.strip()

    # Filter out full sentences (anything with more than 6 words or certain keywords)
    word_count = len(outlet.split())
    if word_count > 6 or any(keyword in outlet.lower() for keyword in ['i listen', 'i occasionally', 'i also', 'i avoid', 'i sometimes']):
        return None

    # Normalize common variations
    outlet_lower = outlet.lower()

    # Dictionary of variations -> standard name
    normalization_map = {
        'cnn': 'CNN',
        'foxnews': 'Fox News',
        'fox news': 'Fox News',
        'npr': 'NPR',
        'national public radio': 'NPR',
        'bbc': 'BBC',
        'bbc news': 'BBC',
        'reuters': 'Reuters',
        'msnbc': 'MSNBC',
        'nbc': 'NBC',
        'abc news': 'ABC News',
        'cbs news': 'CBS News',
        'the washington post': 'The Washington Post',
        'new york times': 'The New York Times',
        'the new york times': 'The New York Times',
        'wsj': 'The Wall Street Journal',
        'the wall street journal': 'The Wall Street Journal',
        'huffington post': 'HuffPost',
        'breitbart': 'Breitbart',
        'politico': 'Politico',
        'the hill': 'The Hill',
        'associated press': 'Associated Press',
        'ap': 'Associated Press',
        'ap news': 'Associated Press',
        'usa today': 'USA Today',
        'google news': 'Google News',
        'google': 'Google News',
        'yahoo': 'Yahoo News',
        'yahoo news': 'Yahoo News',
        'reddit': 'Reddit',
        'twitter': 'Twitter',
        'youtube': 'YouTube',
        'pbs': 'PBS',
        'cnbc': 'CNBC',
        'bloomberg': 'Bloomberg',
        'the guardian': 'The Guardian',
        'the atlantic': 'The Atlantic',
        'the federalist': 'The Federalist',
        'daily mail': 'Daily Mail',
        'dailymail': 'Daily Mail',
        'uk daily mail': 'Daily Mail',
        'one american news': 'One America News',
        'one american news network': 'One America News',
        'oann': 'One America News',
        'drudge report': 'Drudge Report',
        'vice': 'Vice',
        'axios': 'Axios',
        'business insider': 'Business Insider',
        'the intercept': 'The Intercept',
        'alternet': 'AlterNet',
        'the young turks (youtube)': 'The Young Turks',
        'local news': 'Local News',
        'epoch times': 'The Epoch Times',
        'new york post': 'New York Post',
    }

    # Return normalized name or original if not in map
    return normalization_map.get(outlet_lower, outlet.title())

# Step 2: Parse and clean the followed_news_outlets column
parsed_lists = []
for s in data["followed_news_outlets"].to_list():
    if s:
        try:
            raw_list = ast.literal_eval(s)
            # Clean each outlet name and filter out None values
            cleaned_list = [normalize_outlet(outlet) for outlet in raw_list]
            cleaned_list = [x for x in cleaned_list if x is not None]
            parsed_lists.append(cleaned_list)
        except:
            parsed_lists.append([])
    else:
        parsed_lists.append([])

# Step 3: Create a proper List column
data = data.with_columns(
    pl.Series(name="followed_news_list", values=parsed_lists, dtype=pl.List(pl.Utf8))
)

# Step 4: Get all unique outlets (after cleaning)
all_outlets = sorted(list(set([word for sublist in parsed_lists for word in sublist])))

print(f"Found {len(all_outlets)} unique outlets:")
print(all_outlets)

# Step 5: Create one-hot columns
for outlet in all_outlets:
    # Make column names valid (remove spaces and special characters)
    col_name = f"outlet_{outlet.replace(' ', '_').replace('(', '').replace(')', '').replace('-', '_').lower()}"
    data = data.with_columns(
        pl.col("followed_news_list").list.contains(outlet).alias(col_name)
    )

# Step 6: Optional cleanup
data = data.drop(["followed_news_list", "followed_news_outlets"])
````
